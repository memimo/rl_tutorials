```python
import numpy
import matplotlib.pyplot as plt
expected_rewards = numpy.random.random(10)
print expected_rewards
```

```python
def narmed_bandit(arm):
    return numpy.random.random()< expected_rewards[arm]
```

```python
def greedy(num_games = 30, niter=100):
    reward = numpy.zeros((num_games, niter))
    for k in xrange(num_games):
        q = numpy.zeros(10)
        for i in xrange(niter):
            j = numpy.argmax(q)
            reward[k,i] = narmed_bandit(j)
            q[j]+=(reward[k,i] - q[j])/(i+1.)
            
    return reward,q
```

```python
rewards, q = greedy()
plt.plot(rewards.mean(axis=0))
plt.show()
```

```python
def epsilon_greedy(epsilon=0.1,num_games = 30, niter=500):
    reward = numpy.zeros((num_games, niter))
    for k in xrange(num_games):
        q = numpy.zeros(10)
        for i in xrange(niter):
            if numpy.random.binomial(1, epsilon):
                j = numpy.random.randint(10)
            else:
                j = numpy.argmax(q)
            reward[k,i] = narmed_bandit(j)
            q[j]+=(reward[k,i] - q[j])/(i+1.)
            
    return reward,q
```

```python
rewards, q = epsilon_greedy(epsilon=0.3)
plt.plot(rewards.mean(axis=0))
plt.show()
```

```python
def epsilon_greedy2(epsilon=0.1, niter=500):
    reward = numpy.zeros((niter))
    q = numpy.zeros(10)
    for i in xrange(niter):
        if numpy.random.rand() < epsilon:
            j = numpy.random.randint(10)
        else:
            j = numpy.argmax(q)
        reward[i] = narmed_bandit(j)
        q[j]+=(reward[i] - q[j])/(i+1.)
            
    return reward,q
```
